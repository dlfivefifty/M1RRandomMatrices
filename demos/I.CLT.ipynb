```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
p = 0.5
# Generate Bernoulli samples, shifted and scaled to have mean 0 and variance 1
n = 1000
N = 10_000
X = 2*np.random.binomial(1, 0.5, size=(N,n))-1
X
```

```python
x = np.linspace(-4, 4, 300)
plt.figure(figsize=(8, 5))
plt.hist(X[1], bins=60, alpha=0.7, color='steelblue', label='Scaled Bernoulli sum')
plt.show()
```

```python
S = X.sum(axis=1)/np.sqrt(n)
S
```

```python
x = np.linspace(-4, 4, 300)
plt.figure(figsize=(8, 5))
plt.hist(S, bins=60, density=True, alpha=0.7, color='steelblue', label='Scaled Bernoulli sum')
plt.plot(x, np.exp(-x**2/2) / np.sqrt(2 * np.pi), 'r-', linewidth=2, label='Normal')
plt.title(f'CLT: Scaled Sum of Bernoulli(p={p}) vs Gaussian\n(n={n} trials, {N} samples)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
